```python
import os, sys
import pandas as pd
from pathlib import Path

ROOT = Path.cwd().parent
DATA = Path(os.environ.get("DATA_PERFUME_DIR", ROOT / "data"))
sys.path.insert(0, str(ROOT / "backend"))

def load(name):
    for ext in (".parquet", ".csv"):
        p = DATA / f"{name}{ext}"
        if p.exists():
            return pd.read_parquet(p) if ext == ".parquet" else pd.read_csv(p)
    return None
```

```python
perfumes = load("perfumes")
perfume_notes = load("perfume_notes")

perfumes = perfumes.drop_duplicates(subset=["perfume_id"], keep="first")
print(f"perfumes: {perfumes.shape}, unique ids: {perfumes['perfume_id'].nunique()}")
print(f"perfume_notes: {perfume_notes.shape}")
```
